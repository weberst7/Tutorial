```python
  
#einfacher mit methode d.items. Methode gibt ein Tupel zurück das entsprechend direkt im for.. command zerlegt werden kann

dFlughaefen={"Berlin":"BER","Frankfurt":"FRN"}
dFlughaefen["Köln"]="COL"

for key, value in dFlughaefen.items():# items() gibt Tupel zurück welches entpackt wird in die Variablen key und value
      print(key +": "+ value)
```

```python
file = open("../data/names.csv","r")

names={}# leeres dictionairy
   
for line in file:
    row=line.strip().split(",")#Leerzeichen etfernen,Quellzeile in Liste umwandeln gesplittet mit "," da csv
    if row[0]=="Id": #skip Zeile wenn die erste Spalte/Wert der Liste ="Id" = Header
        continue
        
    name=row[1]    
    count=int(row[5] )
    
    #dictionary befüllen
    if name in names:#name schon im dictionary enthalten
        names[name]=names[name]+count# auf den value des keys "name" zugreifen und neuen value (aktueller value+aktuellem count) setzten
    else:
         names[name]=count
        
   
```

```python
for key, value in names.items():#mit.items Tupel erzeugen und in key,value entpacken
    print(key, value)
```
